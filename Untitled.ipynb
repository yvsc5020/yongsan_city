```python
import olefile
import pandas as pd

f = olefile.OleFileIO('data\\221022.hwp')

encoded_text = f.openstream('PrvText').read()
decoded_text = encoded_text.decode('UTF-16')
decoded_text = decoded_text[72:]

listed_text = decoded_text.split('><')
for idx, text in enumerate(listed_text):
    listed_text[idx] = text.replace(">\r\n<", "")

listed_text[0] = listed_text[0].replace("<", "")
day = listed_text[-1].replace(">", "")
listed_text = listed_text[:-1]

columns = ["집회 일시", "집회 장소(행진로)", "신고 인원", "관할서"]
when = []
where = []
who = []
police = []

when_cnt = 0
where_cnt = 1
who_cnt = 2

for idx, text in enumerate(listed_text):
    if idx == when_cnt:
        when.append(text)
        when_cnt += 4
    elif idx == where_cnt:
        where.append(text)
        where_cnt += 4
    elif idx == who_cnt:
        who.append(text)
        who_cnt += 4
    else:
        police.append(text)

list_df = pd.DataFrame(when, columns=["집회 시간"])
list_df["집회 장소"] = where
list_df["집회 인원"] = who
list_df["관할서"] = police
print(day)
list_df
```
